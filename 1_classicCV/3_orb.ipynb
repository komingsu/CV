ORB : Fast detector 와 BRIEF detector의 Orienation 버전을 합쳐서 local feature detecion 을 수행
</br>
PAPER - https://www.gwylab.com/download/ORB_2012.pdf


- 회전에도 Matching을 잘 수행함

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

src = cv2.imread("data/apple_books.jpg")
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
target = cv2.imread("data/apple.jpg", cv2.IMREAD_GRAYSCALE)

orb = cv2.ORB_create(
    nfeatures=40000,
    scaleFactor=1.2,
    nlevels=8,
    edgeThreshold=31,
    firstLevel=0,
    WTA_K=2,
    scoreType=cv2.ORB_HARRIS_SCORE,
    patchSize=31,
    fastThreshold=20,
)

kp1, des1 = orb.detectAndCompute(gray, None)
kp2, des2 = orb.detectAndCompute(target, None)

# 여기서 주의할 점은 ORB에서는 int8 type 데이터가 아니라 Binary 데이터를 쓰기때문에
# 거리를 계산할때 L2 distance 를 쓰지않고 Binary 거리를 계산합니다.

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True) # crossCheck : 양방향

matches = bf.match(des1, des2)
matches = sorted(matches, key=lambda x: x.distance)

for i in matches[:200]:
    idx = i.queryIdx
    x1, y1 = kp1[idx].pt
    cv2.circle(src, (int(x1), int(y1)), 3, (255, 0, 0), 3)
    
    x2, y2 = kp2[idx].pt
    cv2.circle(target, (int(x2), int(y2)), 3, (255, 0, 0), 3)


plt.imshow(src[: ,:, ::-1])
plt.show()

plt.imshow(target[: ,:], 'gray')
plt.show()
```
